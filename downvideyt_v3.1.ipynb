```python
import csv
from pytube import YouTube

def download_youtube_videos_from_csv(csv_file, output_path):
    with open(csv_file, 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip header row if present
        
        for row in csv_reader:
            video_url = row[0]
            try:
                YouTube(video_url).streams.first().download(output_path)
                print(f"Downloaded: {video_url}")
            except Exception as e:
                print(f"Error downloading {video_url}: {str(e)}")

# Example usage
csv_file = 'links.csv'  # Path to the CSV file containing video URLs
output_path = 'fvalue'  # Output directory

download_youtube_videos_from_csv(csv_file, output_path)
```
